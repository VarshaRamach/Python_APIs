```python
# Import dependencies  

import pandas as pd
from citipy import citipy
import numpy as np
import requests as req
import matplotlib.pyplot as plt
import config
from config import api_key
```

```python
# Generate Random 500 latitude and longitude

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
```

```python
# Initialize empty lists 

city_names = []
country_code = []
cloudcover = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
```

```python
# Set up a for loop for 500 random samples

count = 0

for i in range(0,500):
    
    city = citipy.nearest_city(lats[i],lngs[i])
    
    # Append the city and country names to the empty lists 
    
    city_names.append(city.city_name)
    country_code.append((city.country_code).upper())
    
    # Set up the params for the base url
    
    params = {
    'lat': lats[i],
    'lon': lngs[i],
    'units': 'imperial',
    'APPID': api_key
    }
    
    # Set up the base url 
    
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    response = req.get(base_url, params = params)
    
    response_number = [response.url]
    count = count + len(response_number)
        
    city_weather = response.json()
    
    # Fill up the empty list based on API response from the url 
    
    cloudcover.append(city_weather['clouds']['all'])
    date.append(city_weather['dt'])
    humidity.append(city_weather['main']['humidity'])
       
    lat.append(city_weather['coord']['lat'])
    lng.append(city_weather['coord']['lon'])
    
    max_temp.append(city_weather['main']['temp_max'])
    wind_speed.append(city_weather['wind']['speed'])
    
     #lat.append(round(lats[i]))
    #lng.append(round(lngs[i]))
```

```python
# Number of Cities in the loop
count
```

```python
# Names of the cities and countries from the random data loop 

for j in range(len(city_names)):
    print(city_names[j] + ',' + country_code[j])
```

```python
# Scatter plot: Latitude vs Temperature 

plt.scatter(lat, max_temp, marker = 'o', color = 'b', alpha = 0.45, edgecolors="black")

plt.title('Temperature(F) vs Latitude - 8/25/2018 - 12PM')
plt.xlabel('Latitude')
plt.ylabel('Temperature (Farenheit)')
plt.xlim(-100, 100)

plt.savefig('temperature-vs-latitude.png')
plt.show()
```

```python
# Scatter Plot: Latitude vs Humidity

plt.scatter(lat, humidity, marker = 'o', color = 'b', alpha = 0.55, edgecolors="black")

plt.title('Humidity (%) vs Latitude - 8/25/2018 - 12PM')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%))')
plt.xlim(-100, 100)

plt.savefig('humidity-vs-latitude.png')
plt.show()
```

```python
# Scatter Plot: Latitude vs Cloudiness

plt.scatter(lat, cloudcover, marker = 'o', color = 'b', alpha = 0.55, edgecolors="black")

plt.title('Cloud Cover(%) vs Latitude - 8/25/2018 - 12PM')
plt.xlabel('Latitude')
plt.ylabel('Cloud Cover (%))')
plt.xlim(-100, 100)

plt.savefig('cloud_cover-vs-latitude.png')
plt.show()
```

```python
# Scatter Plot: Latitude vs wind speed

plt.scatter(lat, wind_speed, marker = 'o', color = 'b', alpha = 0.50, edgecolors="black")

plt.title('Wind Speed(mph) vs Latitude - 8/25/2018 - 12PM')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.xlim(-100, 100)

plt.savefig('wind_speed-vs-latitude.png')
plt.show()
```

```python
# creating dataframe

cities_weather_df = {'City' : city_names,
              'Temp (F)' : max_temp,
              'Humidity (%)' : humidity,
              'Cloud Cover (%)' : cloudcover,
              'Wind Speed (mph)' : wind_speed,
              'Latitude' : lat,
              'Longitude' : lng
}

cities_weather_df = pd.DataFrame(cities_weather_df)

# set index
cities_weather_df = cities_weather_df.set_index('City')

# write to csv
cities_weather_df.to_csv('cities_weather_data.csv')
```
